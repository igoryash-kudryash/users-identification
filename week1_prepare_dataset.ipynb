```python
def prepare_train_set_opt(path_to_csv_files, session_length=10):
    
    csv_files = glob(path_to_csv_files + "/user????.csv")
    site_amount_dict = {}
    site_index = 1
    
    df_columns = ["site" + str(i) for i in np.arange(1, session_length + 1)]
    df_columns.extend(["user_id"])
    
    substr = "user"
    
    dataframe = pd.DataFrame(columns=df_columns)
    
    all_site_inds = []
    site_inds = np.zeros(session_length + 1, dtype=int)
    
    session_counter = 0
    
    for csv_file in csv_files:
        
        data = pd.read_csv(csv_file, header=0, names=["Date_time", "Sites"])
        start = csv_file.find(substr, len(path_to_csv_files)) + len(substr)
        end = start + 4
        user_id = int(csv_file[start:end:])
        
        
        for i in range(len(data["Sites"])):
            
            unique_site = data["Sites"][i]
            
            if site_amount_dict.get(unique_site) == None:
                site_amount_dict.update({unique_site: (site_index, 1)})
                site_inds[i % session_length] = site_index
                site_index += 1
                
            else:
                ind, count = site_amount_dict.get(unique_site)
                site_amount_dict.update({unique_site: (ind, count + 1)})
                site_inds[i % session_length] = ind
            
            session_counter += 1
            
            if session_counter % session_length == 0:
                site_inds[session_length] = user_id
                all_site_inds.append(site_inds)
                # print(site_inds)
                site_inds = np.zeros(session_length + 1, dtype=int)
                session_counter = 0
                
                
        if session_counter % session_length != 0:
            site_inds[session_length] = user_id
            all_site_inds.append(site_inds)
            #print(site_inds)
            site_inds = np.zeros(session_length + 1, dtype=int)
            session_counter = 0

    dataframe = pd.DataFrame(all_site_inds, columns=df_columns)
    
    
    return dataframe, site_amount_dict
```

```python
def make_csr_matrix(data, session_length=10): 
    # –ù–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏ —Å—Ç—Ä–æ–∫–∏  ùëñ  –∏ —Å—Ç–æ–ª–±—Ü–∞  ùëó  –±—É–¥–µ—Ç —Å—Ç–æ—è—Ç—å —á–∏—Å–ª–æ  ùëõ_ùëñùëó
    # ‚Äì c–∫–æ–ª—å–∫–æ —Ä–∞–∑ —Å–∞–π—Ç  ùëó  –≤—Å—Ç—Ä–µ—Ç–∏–ª—Å—è –≤ —Å–µ—Å—Å–∏–∏ –Ω–æ–º–µ—Ä  ùëñ
    indptr = [0]
    indices = []
    prematrix = []
    vocabulary = {}
    for one_session in data:
        for site_id in one_session:
            index = vocabulary.setdefault(site_id, site_id)
            indices.append(index)
            prematrix.append(1)
        indptr.append(len(indices))
    return csr_matrix((prematrix, indices, indptr), dtype=int)
        
```
